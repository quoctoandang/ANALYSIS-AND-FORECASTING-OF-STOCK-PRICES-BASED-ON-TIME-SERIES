```python
import re
import requests
import pandas as pd
from eodhd import APIClient
from eod import EodHistoricalData
from datetime import datetime
import time
from datetime import datetime, timedelta
```

```python
api_key = '674a03ab53ff16.11596470'
api = APIClient(api_key)
```

```python
_date = "2020-01-01"
to_date = "2024-08-01"
dfs = []
while True:
    resp = api.financial_news(s="TSLA.US", from_date=_date, to_date=to_date, limit=1000)
    df = pd.DataFrame(resp)
    if df.empty or 'date' not in df.columns:
        print("No data returned or 'date' column missing. Exiting loop.")
        break

    dfs.append(df)

    min_date = df['date'].min()
    date_obj = datetime.fromisoformat(min_date)
    date_only = date_obj.date()

    if date_only == datetime.strptime(_date, '%Y-%m-%d').date():
        break


    to_date = str(date_only - timedelta(days=1))
    print("---to_date---: ", to_date)

final_df = pd.concat(dfs, axis=0, ignore_index=True)
```
