```python
import pandas as pd
import yfinance as yf
```

```python
ticker = 'TSLA'
stock = yf.Ticker(ticker)
financials = stock.financials
print(financials)
```

```python
Net_Income = financials.loc['Net Income']
```

```python
Net_Income
```

```python
Net_Income = Net_Income.dropna()
```

```python
Net_Income['2024-12-31'] = Net_Income['2023-12-31']
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TSLA_Technical.csv")
df.head()
```

```python
df['year'].value_counts()
```

```python
Net_Income.index.year
```

```python
df_selected = df[['date', 'close', 'high', 'low', 'open', 'volume', 'year']]
```

```python
df_selected.head()
```

```python
df_selected['net_income'] = df_selected['year'].map(lambda x: Net_Income.loc[Net_Income.index.year == x].values[0] if x in Net_Income.index.year else None)
```

```python
df_selected.head()
```

```python
df_selected['net_income']  = df_selected['net_income'] / 365
```

```python
df_selected['esp'] = df_selected['net_income'] / df_selected['volume']
```

```python
df_selected.head()
```

```python
df_selected['P/E'] = df_selected['close'] / df_selected['esp']
```

```python
df_selected.head()
```

```python
df_selected = df_selected.drop(columns=['net_income', 'year'])
```

```python
df_selected.to_csv("/content/drive/MyDrive/TSLA_fundamental.csv", index=False)
```
